```python
mod = smf.ols(formula='mpg ~ weight', data=df)

res = mod.fit()

res.summary()
```

```python
mod2 = smf.ols(formula='mpg ~ weight + cylinders', data=df)

res2 = mod2.fit()

res2.summary()
```

```python
df
```

```python
mod3 = smf.ols(formula='mpg ~ weight + displacement + origin + model_year + acceleration', data=df)

res3 = mod3.fit()

res3.summary()
```

### Cas pratique pour verifier notre model

```python
# je prend un model en exemple --> chevrolet chevelle malibu
df.head(100)
```

```python
# j'affiche le coef des parametres de mon model

res3.params[['weight', 'displacement', 'origin', 'model_year', 'acceleration']]
```

```python
# je multiplie mes coef avec les valeur de mon dataframe pour l'exemple --> chevrolet chevelle malibu

result = res3.params['weight']*3504 + res3.params['displacement']*307 + res3.params['origin']*1 + res3.params['model_year']*70 + res3.params['acceleration']*12+ res3.params['Intercept']
print(f' Le miles par gallon pour la chevrolet malibu est \n {result}')
```

```python
#result2 = res3.params['weight']*3504 + res3.params['displacement']*307 + res3.params['origin']*1 + res3.params['acceleration']*12+ res3.params['Intercept']
#print(f' La prédiction pour la chevrolet malibu est \n {result2}')
```

```python
def prediction_mpg(weight, displacement, origin, model_year, acceleration, name):
    result = res3.params['weight']* weight + res3.params['displacement']* displacement + res3.params['origin']* origin + res3.params['model_year']*model_year + res3.params['acceleration']*acceleration + res3.params['Intercept']
    result_min = round((res3.rsquared * result) ,2)
    result_max = round(((1 - res3.rsquared) * result + result),2)
    return f'Votre voiture de marque {name} consomme entre {result_min} et {result_max} galons par mile'
    
    
```

```python
print(prediction_mpg(weight=2945, displacement=232.0, origin=1, model_year=73, acceleration=16.0, name='subaru'))
```

```python
print(prediction_mpg(weight=2945, displacement=232.0, origin=1, model_year=73, acceleration=16.0, name=''))
```
